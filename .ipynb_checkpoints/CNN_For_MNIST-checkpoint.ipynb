```python
import torch 
import torch.nn as nn
import torch.optim as optim

import torchvision
from torchvision.datasets import MNIST

from torch.utils.data import DataLoader
import torchvision.transforms as transforms
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5), (0.5))
])
```

```python
trainset = MNIST("./cnn", train=True, download=False, transform  = transform)
testset = MNIST("./cnn", train=False, download=False, transform = transform)
```

```python
trainloader = DataLoader(trainset, batch_size=64, shuffle=True)
testloader = DataLoader(testset, batch_size=64)
```

# Build CNN Module For MNIST Dataset

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv_Layers = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),

            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),

            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2,2)  
        )
        self.FLC_Layers = nn.Sequential(
            nn.Linear(3*3*128, 256),
            nn.ReLU(),
            
            nn.Linear(256, 10)
        )
    def forward(self, x):
        x = self.conv_Layers(x)
        x = x.view(x.size(0), -1) # Fallten 
        x = self.FLC_Layers(x)

        return x
```

```python
model = CNN()
criterion = nn.CrossEntropyLoss()
optimizers = optim.Adam(model.parameters())
```

```python
epochs = 10
train_loss = []
test_loss = []
correct = 0
total = 0

for epoch in range(epochs):
    model.train()
    running_loss = 0.0

    for images, labels in trainloader:
        optimizers.zero_grad()
        output = model(images)
        loss = criterion(output, labels)

        loss.backward()
        optimizers.step()
        running_loss += loss

    per_train = running_loss / len(trainloader)

    train_loss.append(per_train)


    model.eval()
    val_loss = 0.0
    with torch.no_grad():
        for images, labels in testloader:
            outputs = model(images)
            loss = criterion(outputs, labels)

            val_loss += loss

        per_val = running_loss / len(testloader)
        test_loss.append(per_val)


        print(f"In epoch {epoch + 1} / {epochs}, the training loss is {per_train} and validation loss is {per_val}.")

            _,predict = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predict == labels).sum().item()

        

        print("Correct: ", correct)
        print("Total: ", total)
            

        
```
